```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset.columns
```

```python
dataset["classification"].value_counts()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset['classification_yes'].value_counts()
```

```python
dataset.columns
```

```python
independant=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependant=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={'kernel':['linear','rbf','poly','sigmoid'],
            'gamma':['auto','scale'],
            'C':[10,100,1000,2000,3000]}

grid=GridSearchCV(SVC(probability=True),param_grid,refit=True,verbose=3,cv=5,n_jobs=-1,scoring='f1_weighted')

grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
print(re)
```

```python
grid_predictions=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,grid_predictions)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,grid_predictions)
```
